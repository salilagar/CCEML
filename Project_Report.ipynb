# Scope:
 In this project we have taken the standard MNIST database of handwritten digits (0-9) and used a CNN and MLP model to tran it. The model is validated against the MNIST provided test dataset itself. We allow the user of this project to load his own handwritten digits and use the trained model to recognize the digits. The recognition of digits is also timed to get a view/sense of overall time the model takes for inferencing of handwritten digits.

The idea is to not just compare the training time, but also to achieve similar accuracy levels in two models and compare the inferencing times when the model are shown digits written by a users. 


# How to create your own digits:
MNIST images are provided in grayscale 28x28 pixels. Each image is a handwritten single digit number with black background and white colored number. In order to create similar images we used MSPaint application in Windows. Create a canvas of 28x28 pixels first, zoom to max as otherwise it is too small. Now fill the background with black color. Use a white colored pen to write the digit (0-9) of your choice. Save the file in jpg format and upload in project/data folder in juliabox.

For creating a 2 or more digit number (like 34, 560 etc.) create the pixel canvas as 28xn*28 where n is the number of digits you want to write. Write numbers in each 28x28 box only, take care not to cross the boundary of each box (like we used to write in childhood).

Note: This jpg will be in RGB format and not gray scale.

# Results: 
 In both the trained models, CNN and MLP, the digits are identified properly to a fair extent (~85% - ~95% accuracy). However there are still some digits that get mis-classified. On this count, both MLP and CNN model perform almost equally good. 
 On accuracy front CNN performs better, getting to almost ~95% with lesser amount in training, whereas MLP based model reaches greater than 90% in lesser training time.
 
 However they differ in the time taken in inferencing. MLP based model takes ~100 micro seconds for a single digit identification, the CNN based model takes ~500 micro seconds for identification. This results in 5 time the delay in indentification.
 
We need to note that both the models have a different level of complexity and hence comparision may not best way to analyze the issue. However in many industries the speed at which the inferencing happens for each instance matters a lot e.g. in case of near-real time processes in radio commnunication. Hence choosing the right the kind of model which takes into account inferencing time as well. For example, in case of 5G system there are requirements for Ultra Low Latency (< 1ms) use cases where the latency should be within 1 ms. In such application using Machine Learning inferencing time plays a very critical role along with accuracy.
